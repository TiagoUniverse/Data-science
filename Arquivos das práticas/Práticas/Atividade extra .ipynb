9- Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:
salário bruto.
quanto pagou ao INSS.
quanto pagou ao sindicato.
o salário líquido.
calcule os descontos e o salário líquido

```python
valorHora = float(input("Quanto voce recebe por hora:"))
totalHoras = float(input("Quantas horas voce trabalha por mes:"))
valorTotal = valorHora * totalHoras

##Tratamento das variaveis
inss = (valorTotal * 8) /  100
sindicato = (valorTotal * 5) / 100
imposto = (valorTotal * 11) /  100
salarioLiquido = (valorTotal - sindicato - inss - imposto)
descontoLiquido = salarioLiquido + inss + sindicato + imposto

##Resultados
print("O seu salario e: ", valorTotal)
print("Valor direcionado ao INSS: ", inss)
print("Valor direcionado ao sindicato: ", sindicato)
print("Valor de impoSto de renda: ", imposto)

print("O seu salario liquido e: ", salarioLiquido)
print("Descontos com o salario liquido:" , descontoLiquido)
```

10- Faça um programa que peça o tamanho de um arquivo para download (em MB) e a velocidade de um link de Internet (em Mbps), calcule e informe o tempo aproximado de download do arquivo usando este link (em minutos).

```python
tamanhoArquivo = float(input("Me informe o tamanho de um arquivo pra download (em MB): "))
velocidadeInternet = float(input("Me informe agora sua velocidade de conexão de internet (em Mbps): "))

tam_bits = tamanhoArquivo * 8 * 1024 * 1024 #Pra ser mais simples, melhor transformar em bytes (1024 duas vezes porque passa de MB por KB e dps B)
tempoDownload = tam_bits / (velocidadeInternet * 1024 * 1024) #Mesma lógica para velocidade de internet Mbps > Kbps > Bps
tempoMinutos = tempoDownload / 60

print("O tempo aproximado de download é de", round(tempoMinutos), "minutos.") #round pra arredondar o float
```
