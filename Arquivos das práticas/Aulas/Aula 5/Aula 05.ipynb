# Introdução

## Importação

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# O que sabemos?
empresas = ['Apple', 'Google', 'Amazon' ,
            'Microsoft', 'Tencent', "McDonald's" ,
            'Visa', 'Facebook', 'Alibaba', 'Louis Vuitton']
```

```python
 bilhoes = [947.062, 819.573, 705.646,
            611.460, 214.023, 196.526,
            191.032, 186.421, 169.966,
              124.273]
```

```python
# Já é possível plotar, só com as variáveis
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
```

# Tamanho da figura

```python
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
plt.show()
```

```python
# Redimencionando
plt.rcParams["figure.figsize"] = (15,8)
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
```

```python
# Conhecendo o fontsize
plt.plot(empresas, bilhoes)
plt.title("Empresas mais ricas do mundo", fontsize=25)
plt.xlabel("Empresas", fontsize=20)
plt.ylabel("Bilhões", fontsize= 20)
```

```python
# Mudando a core legendado
plt.plot(empresas, bilhoes, color = 'green')
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
plt.legend(['Bilhões de dólares'] , loc = 'right' , fontsize=15)
```

```python
# Aumentar a espessura da linha lw
plt.plot(empresas, bilhoes, color = 'green' , lw=10)
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
plt.legend(['Bilhões de dólares'] , loc = 'right' , fontsize=15)
```

```python
# Redimencionando marcador
plt.plot(empresas, bilhoes, color = 'green' , marker= 'o')
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
plt.legend(['Bilhões de dólares'] , loc = 'right' , fontsize=15)
```

```python
# Adicionando as linhas de grade
plt.plot(empresas, bilhoes, color = 'green' , marker='o')
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas")
plt.ylabel("Bilhões")
plt.legend(['Bilhões de dólares'] , loc = 'right' , fontsize=15)
plt.grid(True)
```

```python
# Mudando o tipo de gráfico e colocando a legenda acima
plt.bar(empresas, bilhoes, color = 'green')
plt.title("Empresas mais ricas do mundo")
plt.xlabel("Empresas", fontsize=20)
plt.ylabel("Bilhões", fontsize=20)
plt.legend(['Bilhões de dólares'] , loc = 'right' , fontsize=15)
```

```python
# Gráfico mais simples
plt.bar(empresas, bilhoes)
```

# Lista de cores

```python
cores = ['black', 'dimgray', 'dimgrey' , 'gray', 'grey',
         'darkgray' , 'darkgrey', 'silver' , 'lightgray' , 'lightgrey']
```

```python
plt.bar(empresas, bilhoes, color= cores, edgecolor= 'black')
plt.title('Empresas mais ricas do mundo', fontsize= 25)
plt.ylabel('Bilhões', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
cores = ['black', 'pink', 'yellow' , 'gray', 'purple',
         'blue' , 'white', 'green' , 'orange' , 'lightgrey']
```

```python
plt.bar(empresas, bilhoes, color= cores, edgecolor= 'black')
plt.title('Empresas mais ricas do mundo', fontsize= 25)
plt.ylabel('Bilhões', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```
