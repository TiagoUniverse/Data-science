# Análise dos dados de vacina de Recife

Hipótese: Buscar entender se teve mais homens ou mulheres que foram vacinadas

Analisando o caso do feminino. 


No final da pesquisa, ele informa a quantidade de registros com o sexo igual à Feminino

```python
resultadoFem = df1[df1['sexo'] == 'FEMININO']
(resultadoFem)
```

Resultado: 

1941977 registros femininos

Analisando o masculino:

```python
resultadoMasc = df1[df1['sexo'] == 'MASCULINO']
(resultadoMasc)
```

Resultado:

1460003 registros masculinos

Logo, conclui-se que em Recife teve mais vacinação de mulheres registradas no sexo feminino do que dos homens.
