2ª prática de lista

Crie uma lista chamada minhaLista com os seguintes itens: 76, 92.3 , "oi", True, 4, 76. Depois execute os seguintes comandos: 1- Inserir pitomba e 76 no final da lista, Inserir o valor "Cibele" na posição de índice 3 e Inserir o valor 99 no início da lista

```python
minhaLista = [76, 92.3 , "oi" , True, 4, 76]
```

```python
minhaLista.append ("pitombas") # Primeira opção
```

```python
minhaLista[6:] = 'pitombas', 76 # segunda opção
```

```python
minhaLista = minhaLista + ["pitombas"] #terceira opção
```

```python
minhaLista = minhaLista + ["pitombas" , 76]  # Quarta opção
```

```python
minhaLista = minhaLista + ["pitombas"] , [76] # Quinta opção
```

```python
minhaLista.append(76) # Sexta opção
```

```python
minhaLista += minhaLista + ["pitombas"] # Sétima
```

```python
minhaLista
```
